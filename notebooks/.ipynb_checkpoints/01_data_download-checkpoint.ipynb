# 01 — Dataset Preparation (local MacBook)

Extracts the three manually-downloaded archives in `raw/` into `data/` and pulls HAM10000 from Kaggle (or from a local copy if you pre-downloaded it).

Expected layout before running this notebook:

```
raw/
  release_v0.zip                    # Derm7pt (from https://derm.cs.sfu.ca)
  ddidiversedermatologyimages.zip   # DDI    (from Stanford AIMI)
  fitzpatrick17k.csv                # Fitzpatrick17k annotations
```

````python
import os, sys, logging
from pathlib import Path

# Resolve project root regardless of where the notebook is launched from.
PROJECT_ROOT = Path(os.getcwd())
if PROJECT_ROOT.name == 'notebooks':
    PROJECT_ROOT = PROJECT_ROOT.parent
sys.path.insert(0, str(PROJECT_ROOT))
os.chdir(PROJECT_ROOT)

logging.basicConfig(level=logging.INFO, format='%(levelname)s %(name)s %(message)s')
print('PROJECT_ROOT:', PROJECT_ROOT)
````

````python
import yaml
from src.data import download
from src.utils import describe_device, get_device

cfg = yaml.safe_load((PROJECT_ROOT / 'config.yaml').read_text())
paths = cfg['paths']['datasets']
raw_dir = Path(cfg['paths'].get('raw_root', 'raw'))
print('Using device:', describe_device(get_device()))
print('Raw dir     :', raw_dir.resolve())
for k, v in paths.items():
    print(f'  {k:16s} -> {v}')
````

````python
# Sanity-check what's in raw/
if raw_dir.exists():
    for f in sorted(raw_dir.iterdir()):
        size_mb = f.stat().st_size / (1 << 20) if f.is_file() else 0
        kind = 'file' if f.is_file() else 'dir '
        print(f'  {kind}  {f.name:40s}  {size_mb:9.2f} MB')
else:
    print('raw/ does not exist. Create it and drop the three archives in.')
````

````python
# HAM10000 (tries raw/ first, then kagglehub).
try:
    download.download_ham10000(paths['ham10000'], raw_dir=raw_dir)
except Exception as exc:
    print('HAM10000 failed:', exc)
    print('Either drop the Kaggle ZIP at raw/ham10000.zip or run `pip install kagglehub`')
````

````python
# Derm7pt (uses raw/release_v0.zip if present).
download.download_derm7pt(paths['derm7pt'], raw_dir=raw_dir)
````

````python
# Fitzpatrick17k: copy CSV + (optional) scrape images by URL.
# Set FITZ17K_FETCH_IMAGES=1 to actually download the 16k images (slow,
# ~30-60 min, many broken URLs). Otherwise only the CSV is installed and
# the dataset loader will drop rows whose image isn't on disk.
fetch_images = os.environ.get('FITZ17K_FETCH_IMAGES', '0') == '1'
download.download_fitzpatrick17k(
    paths['fitzpatrick17k'],
    raw_dir=raw_dir,
    images_url=os.environ.get('FITZ17K_IMAGES_URL'),
    fetch_images_from_urls=fetch_images,
    max_images=int(os.environ.get('FITZ17K_MAX_IMAGES', 0)) or None,
)
````

````python
# DDI (uses raw/ddidiversedermatologyimages.zip if present).
download.download_ddi(paths['ddi'], raw_dir=raw_dir)
````

````python
# Readiness summary: build each Dataset and report N, class balance,
# and (for Fitz17k / DDI) bucket distribution.
import pandas as pd
from src.data.datasets import build_dataset
from src.data.transforms import biomedclip_eval_transform

tfm = biomedclip_eval_transform(224)
summary_rows = []
for name in ['ham10000', 'derm7pt', 'fitzpatrick17k', 'ddi']:
    try:
        ds = build_dataset(name, root=paths[name], transform=tfm)
    except Exception as exc:
        summary_rows.append({'dataset': name, 'n': 0, 'error': str(exc)[:60]})
        continue
    n = len(ds)
    if n == 0:
        summary_rows.append({'dataset': name, 'n': 0, 'error': 'empty'})
        continue
    # Sample 256 rows for quick stats
    sample_idx = list(range(0, min(n, 256)))
    recs = [ds._row_to_record(ds.df.iloc[i]) for i in sample_idx]
    pos_rate = sum(r['diagnosis'] for r in recs) / len(recs)
    buckets = pd.Series([r['fitzpatrick_bucket'] for r in recs]).value_counts(dropna=False).to_dict()
    summary_rows.append({
        'dataset': name,
        'n': n,
        'pos_rate_sample': round(pos_rate, 3),
        'bucket_sample': buckets,
    })
pd.DataFrame(summary_rows)
````
